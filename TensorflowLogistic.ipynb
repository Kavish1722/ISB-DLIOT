```python
import numpy as np
import tensorflow as tf
import pandas
# Model parameters

A = tf.Variable(tf.random_normal(shape=[4, 1]))
b = tf.Variable(tf.random_normal(shape=[1, 1]))

# Model input and output
data = tf.placeholder(dtype=tf.float32,shape=[None,4])
target = tf.placeholder(dtype=tf.float32,shape=[None,1])

# Declare the model 
mod = tf.matmul(data, A) + b


#loss function
#sigmoid cross-entropy
loss = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(logits=mod, labels=target))


# Define the accuracy
prediction = tf.round(tf.sigmoid(mod))
# Bool into float32 type
correct = tf.cast(tf.equal(prediction, target), dtype=tf.float32)
# Average
accuracy = tf.reduce_mean(correct)

#learning rate etc.
learning_rate = 0.01
iter_num = 1500

# optimizer
optimizer = tf.train.GradientDescentOptimizer(learning_rate)

goal = optimizer.minimize(loss)
```

```python
train = pandas.read_csv("/Users/kavish/Desktop/DLIOT/iris_training.csv")
test = pandas.read_csv("/Users/kavish/Desktop/DLIOT/iris_test.csv")
train_X = train.loc[:,['instances','features','versicolor','setosa']]
train_y = train.loc[:,['virginica']]
test_X = test.loc[:,['instances','features','versicolor','setosa']]
test_y = test.loc[:,['virginica']]
print(train_X)
print(train_y.shape)
init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)
for epoch in range(iter_num):
    sess.run(goal, {data: train_X, target: train_y})
    temp_loss = sess.run(loss, {data: train_X, target: train_y})
    # convert into a matrix, and the shape of the placeholder to correspond
    temp_train_acc = sess.run(accuracy, {data: train_X, target:train_y})
    temp_test_acc = sess.run(accuracy, {data: test_X, target:test_y})
    # output
    if (epoch + 1) % 300 == 0:
        print('epoch: {:4d} loss: {:5f} train_acc: {:5f} test_acc: {:5f}'.format(epoch + 1, temp_loss,
                                                                          temp_train_acc, temp_test_acc))
import matplotlib.pyplot as plt   
plt.plot(temp_train_acc, 'b-', label='train accuracy')
plt.plot(temp_test_acc, 'k-', label='test accuracy')
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.title('Train and Test Accuracy')
plt.legend(loc='best')
plt.show()
```
